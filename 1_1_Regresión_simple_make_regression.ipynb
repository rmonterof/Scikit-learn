# Regresi√≥n lineal simple

Ejemplo con datos ficticios generados con ``datasets.make_regression``

```python
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
plt.style.use('seaborn-v0_8-whitegrid')
plt.rcParams['figure.dpi'] = 110
```

```python
X, y = make_regression(n_features=1, n_samples=500, random_state=42, noise=10)
plt.xlabel('Feature X')
plt.ylabel('Target y')
plt.scatter(X, y, s=5)
```

```python
print(X.shape)
print(X[:15])
```

```python
print(y.shape)
print(y[:15])
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
y_pred = model.predict(X)
y_pred
```

```python
plt.scatter(X, y, s=10, label='training')
plt.plot(X, y_pred, 'r', label='predictions')
plt.xlabel('Feature X')
plt.ylabel('Target y')
plt.legend()
```
